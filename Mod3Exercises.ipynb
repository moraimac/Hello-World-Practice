### Exercise 7:

Using the dataframe from the previous exercise, pick a method (Standard Deviation or Interquartile Range) and remove the outliers from the "FuelEconomy" column.

```python
 #drop rows with any missing data
travel_df_no_missing = travel_df.dropna()
travel_df_no_missing
```

```python
#replacing values with new value
travel_df_no_missing.loc[travel_df_no_missing['FuelEconomy']== '-', 'FuelEconomy']=0
```

```python
travel_df_no_missing.head()
```

```python
travel_df_no_missing['FuelEconomy'] = travel_df_no_missing['FuelEconomy'].astype(float)
```

```python
#Interquartile Range Method

q1 = travel_df_no_missing['FuelEconomy'].quantile(.25)
q3 = travel_df_no_missing['FuelEconomy'].quantile(.75)
iqr = q3-q1
toprange = q3 + iqr * 1.5
botrange = q1 - iqr * 1.5

newdf = travel_df_no_missing.copy()
newdf = newdf.drop(newdf[newdf['FuelEconomy'] > toprange].index)
newdf = newdf.drop(newdf[newdf['FuelEconomy'] < botrange].index)

newdf.head()
newdf.head()
```
